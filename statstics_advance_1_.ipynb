1. What is a random variable in probability theory?
- A random variable is a function that assigns numerical values to the outcomes of a random experiment. It maps each possible outcome in the sample space to a real number, and its distribution describes the probabilities of these outcomes

2. What are the types of random variables?
Random variables are classified into:

- Discrete Random Variables: Take countable values (e.g., number of heads in coin tosses).

- Continuous Random Variables: Take any value within a range or interval (e.g., height or weight)

3. What is the difference between discrete and continuous distributions?
- Discrete Distributions: Deal with countable outcomes (e.g., binomial, Poisson distributions). Probabilities are assigned to specific values.

- Continuous Distributions: Deal with uncountable outcomes within intervals. Probabilities are calculated as areas under a curve using a probability density function (PDF)

4. What are probability distribution functions (PDF)?
- A PDF is used for continuous random variables to describe the likelihood of the variable falling within a specific range. It represents probability density rather than absolute probabilities

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- PDF: Shows the likelihood of a random variable taking values close to a specific point.

- CDF: Represents the cumulative probability that the random variable is less than or equal to a certain value

6. What is a discrete uniform distribution?
- A discrete uniform distribution assigns equal probability to all distinct outcomes in a finite set. For example, rolling a fair die results in equal probabilities for numbers 1 through 6

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of population shape. This enables statistical inference for large samples

- The CLT explains why many real-world phenomena exhibit normal distributions when aggregated from smaller independent components

- Z-statistics compare sample data against population parameters under known standard deviations. They are used for testing hypotheses about means or proportions

17. What are point estimates and interval estimates in statistics?
- Point Estimates: Single value estimates for population parameters (e.g., sample mean).

- Interval Estimates: Range within which population parameters likely fall, often expressed as confidence intervals

18. What is the significance of confidence intervals in statistical analysis?
- Confidence intervals provide ranges for population parameters with specified probabilities (e.g., 95%). They quantify uncertainty in estimates

- Z-scores determine critical values for confidence intervals based on desired confidence levels (e.g., 1.96 for 95% confidence)

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt

n = 6
outcomes = np.arange(1, n + 1)
probabilities = [1/n] * n
plt.bar(outcomes, probabilities, color='skyblue', edgecolor='black')
plt.title("Discrete Uniform Distribution PMF")
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.show()
```

```python
# 3. Calculate the PDF of a Bernoulli distribution
def bernoulli_pdf(p, x):
    if x not in [0, 1]:
        return 0
    return p**x * (1 - p)**(1 - x)

p = 0.7
x = 1
print(bernoulli_pdf(p, x))
```

```python
# 4. Simulate a binomial distribution and plot its histogram
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 10
p = 0.5

data = binom.rvs(n=n, p=p, size=1000)

plt.hist(data, bins=range(n+2), density=True, alpha=0.7, color='blue', edgecolor='black', align='left')
plt.title("Binomial Distribution Histogram")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# 6. Calculate and plot the CDF of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt

n = 6
outcomes = np.arange(1, n + 1)
probabilities = [1/n] * n

cdf = np.cumsum(probabilities)

plt.step(outcomes, cdf, where='post', color='green')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Outcomes")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python
9. Calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

data = np.random.normal(loc=50, scale=10, size=100)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=20, density=True, alpha=0.7, color='coral', edgecolor='black')
plt.title("Z-Scores Histogram")
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.show()
```

```python
# 10. Implement the Central Limit Theorem for a non-normal distribution
import numpy as np
import matplotlib.pyplot as plt

population_data = np.random.exponential(scale=1.0, size=100000)

sample_sizes = [5, 30, 50]
sample_means = []

for size in sample_sizes:
    means = [np.mean(np.random.choice(population_data, size=size)) for _ in range(1000)]
    sample_means.append(means)

for i, size in enumerate(sample_sizes):
    plt.hist(sample_means[i], bins=30, density=True, alpha=0.7,
             label=f"Sample Size {size}", edgecolor='black')

plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# 16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, loc=0, scale=1)
    plt.plot(x, y, color='blue')
    plt.title("Standard Normal Distribution")
    plt.xlabel("Z-Score")
    plt.ylabel("Density")
    plt.show()

plot_standard_normal()
```

```python
# 19. Implement hypothesis testing using Z-statistics for a sample dataset
from scipy.stats import norm

sample_mean = 52
population_mean = 50
std_dev = 5
n = 30

z_statistic = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

print(f"Z-Statistic: {z_statistic}")
print(f"P-Value: {p_value}")
```

```python
# 20. Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import norm

data = np.random.normal(loc=50, scale=10, size=100)
mean = np.mean(data)
std_error = np.std(data) / np.sqrt(len(data))
confidence_level = 0.95

z_critical = norm.ppf((1 + confidence_level) / 2)
margin_of_error = z_critical * std_error

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
# 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
data_normal = np.random.normal(loc=50, scale=10, size=1000)

range_value = np.ptp(data_normal)
q1, q3 = np.percentile(data_normal, [25, 75])
iqr_value = q3 - q1

print(f"Range: {range_value}")
print(f"IQR: {iqr_value}")
```

```python
# 27. Implement Z-score normalization on a dataset and visualize its transformation
data_original = np.random.normal(loc=50, scale=10, size=1000)

mean_original = np.mean(data_original)
std_original = np.std(data_original)

z_scores_transformed_data = (data_original - mean_original) / std_original

plt.hist(z_scores_transformed_data, bins=30)
plt.title("Z-Score Normalized Data")
plt.show()
```
